```python
wawa = first_twenty_songs[0]['artists']
```

```python
first_twenty_songs[0].keys()
```

```python
# Your code here:
song_dictionary = {}

for song in first_twenty_songs:
    artist_name_list = []
    artists = song['artists']
    for artist in artists:
        artist_name_list.append(artist['name'])
    
    artist_name_tuple = tuple(artist_name_list)
    
    song_dictionary.update({song['name']: artist_name_tuple})
    
song_dictionary
```
